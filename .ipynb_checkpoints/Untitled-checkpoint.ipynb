```python
import pandas as pd
df = pd.read_csv("data/ip-sama.csv")

counts = df["ID"].value_counts()
counts.head(10).plot(kind="bar")
```

```python
import matplotlib.pyplot as plt

counts.head(5).plot(kind="pie", autopct="%1.1f%%")
plt.gca().set_ylabel("")
centre_circle = plt.Circle((0,0),0.70)
plt.gca().add_artist(centre_circle)
plt.show()
```
